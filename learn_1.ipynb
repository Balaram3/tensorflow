```python
import numpy as np
import matplotlib.pyplot as plt
```

Preparing Data for Model

```python

x_data=[1.0,2.,3.]
y_data=[2.0,4.0,6.0]

w=0.0

```

Defining forward function and loss

```python
def fitter(x):
    return x*w
def loss(x,y):
    y_pred=fitter(x)
    return (y-y_pred)**2

```

Training to get desired Model 

```python

w_list=[]
l_list=[]



for w in np.arange(0,4.3,0.025):
    print("w=",w)
    l_sum=0
    for x_val,y_val in zip(x_data,y_data):
        l_sum+=loss(x_val,y_val)
        print(l_sum/3)
    w_list.append(w)
    l_list.append(l_sum/3)

plt.plot(w_list,l_list)
plt.show()
```
